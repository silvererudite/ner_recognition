```python
def create_sequences(data):
    seqs = []
    inter_arr = []
    for i, elem in enumerate(data.values):
        #print(ent)
        if elem != 0:
            inter_arr.append(elem)
        else:
            seqs.append(inter_arr)
            inter_arr = []
    return seqs
```

```python
train_entities = create_sequences(train_data['entity'])
train_tags = create_sequences(train_data['tag'])
valid_entities = create_sequences(valid_data['entity'])
valid_tags = create_sequences(valid_data['tag'])
```

```python
train_sequence = pd.DataFrame(list(zip(train_entities, train_tags)),
               columns =['entities', 'tags'])
```

```python
valid_sequence = pd.DataFrame(list(zip(valid_entities, valid_tags)),
               columns =['entities', 'tags'])
```

```python
train_sequence.to_csv('train_sequence.csv', header=True, index=False)
valid_sequence.to_csv('valid_sequence.csv', header=True, index=False)
```
